```python
import numpy as np
from keras.models import Sequential
from keras.layers import Dense, Dropout, LSTM
from keras.callbacks import LambdaCallback
```

```python
file_name = "sherlock_holmes.txt"  # Make sure the uploaded file has this name
text = open(file_name, 'r', encoding='utf-8').read().lower()
```

```python
# Create a list of unique characters
chars = sorted(list(set(text)))
chars_to_numbers = {c: i for i, c in enumerate(chars)}
numbers_to_chars = {i: c for i, c in enumerate(chars)}

# Split text into sequences
sequence_length = 40
training_sequences = []
training_sequences_next_character = []

for i in range(0, len(text) - sequence_length, 40):
    training_sequences.append(text[i: i + sequence_length])
    training_sequences_next_character.append(text[i + sequence_length])

# One-hot encode sequences
X = np.zeros((len(training_sequences), sequence_length, len(chars)), dtype=bool)
y = np.zeros((len(training_sequences), len(chars)), dtype=bool)

for i, sentence in enumerate(training_sequences):
    for t, char in enumerate(sentence):
        X[i, t, chars_to_numbers[char]] = 1
    y[i, chars_to_numbers[training_sequences_next_character[i]]] = 1



```

```python
# Function to generate new text after each epoch
def generate_new_text(epoch, _):
    seed_text = "when you have eliminated the impossible,"  # starting text
    new_text = ""

    for i in range(1000):
        x_pred = np.zeros((1, sequence_length, len(chars)))
        for t, char in enumerate(seed_text):
            if char in chars_to_numbers:
                x_pred[0, t, chars_to_numbers[char]] = 1

        predicted_letter_prob = model.predict(x_pred, verbose=0)[0]

        predicted_letter_prob *= 0.99  # fix floating point precision

        next_index = np.argmax(np.random.multinomial(1, predicted_letter_prob, 1))
        next_char = numbers_to_chars[next_index]

        new_text += next_char
        seed_text = seed_text[1:] + next_char

    print("\n--- Generated Text After Epoch {} ---\n{}".format(epoch + 1, new_text))
    print("\n-----------------------------------\n")
```

```python
# Build the LSTM model
model = Sequential()
model.add(LSTM(128, input_shape=(sequence_length, len(chars)), return_sequences=True))
model.add(LSTM(128))
model.add(Dropout(0.5))
model.add(Dense(len(chars), activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer="adam")
```

```python
# Train the model
model.fit(
    X,
    y,
    batch_size=128,
    epochs=10,  # start with smaller number of epochs for testing
    verbose=2,
    callbacks=[LambdaCallback(on_epoch_end=generate_new_text)]
)
```
